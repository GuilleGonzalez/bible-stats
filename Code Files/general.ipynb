```python
from bible_stats import BibleStats # My Custom Class for the WPM Function & Constants
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

books_df = pd.read_csv('General/Raw Bible Books (General).csv').set_index('Book') # Dependent on File "General/Raw Bible Books (General).csv"
sections_df = pd.read_csv('General/Raw Bible Sections (General).csv').set_index('Section') # Dependent on File "General/Raw Bible Sections (General).csv"
```

```python
books_df['Vs/Ch'] = round(books_df['Verses'] / books_df['Chapters'], 1)
books_df['Wd/Ch'] = round(books_df['Words'] / books_df['Chapters'], 1)
books_df['Wd/Vs'] = round(books_df['Words'] / books_df['Verses'], 1)

for time in range(BibleStats.MIN_WPM, BibleStats.MAX_WPM, BibleStats.INCREMENT):
    books_df[f'Time ({time} WPM)'] = books_df['Words'].apply(lambda x: BibleStats.wpm(x, time))
books_df.to_csv('General/Bible Books (General).csv')
books_df
```

```python
sections_df['Ch/Book'] = round(sections_df['Chapters'] / sections_df['Books'], 1)
sections_df['Vs/Book'] = round(sections_df['Verses'] / sections_df['Books'], 1)
sections_df['Wd/Book'] = round(sections_df['Words'] / sections_df['Books']).astype('int')

sections_df['Vs/Ch'] = round(sections_df['Verses'] / sections_df['Chapters'], 1)
sections_df['Wd/Ch'] = round(sections_df['Words'] / sections_df['Chapters'], 1)
sections_df['Wd/Vs'] = round(sections_df['Words'] / sections_df['Verses'], 1)

for time in range(BibleStats.MIN_WPM, BibleStats.MAX_WPM, BibleStats.INCREMENT):
    sections_df[f'Time ({time} WPM)'] = sections_df['Words'].apply(lambda x: BibleStats.wpm(x, time))
sections_df.to_csv('General/Bible Sections (General).csv')
sections_df
```
