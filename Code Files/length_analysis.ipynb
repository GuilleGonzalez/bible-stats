```python
import pandas as pd

VERSION = 'KJV'

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

sections = pd.read_csv(f'Code Files/CSV/{VERSION} Sections.csv').set_index('Section').drop(['Time (150 WPM)', 'Time (200 WPM)', 'Time (250 WPM)'], axis=1)
books = pd.read_csv(f'Code Files/CSV/{VERSION} Books.csv').set_index('Book').drop(['Time (150 WPM)', 'Time (200 WPM)', 'Time (250 WPM)'], axis=1)
chapters = pd.read_csv(f'Code Files/CSV/{VERSION} Chapters.csv').set_index('Reference').drop(['Time (150 WPM)', 'Time (200 WPM)', 'Time (250 WPM)'], axis=1)
verses = pd.read_csv(f'Code Files/CSV/{VERSION} Verses.csv').set_index('Reference')
sections
```
