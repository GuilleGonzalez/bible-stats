```python
import pandas as pd
from bible_stats import BibleStats

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

WPM_SPEED = 300
VERSION = 'KJV'

sections = pd.read_csv(f'CSV/{VERSION} Sections.csv').set_index('Section')
books = pd.read_csv(f'CSV/{VERSION} Books.csv').set_index('Book')
chapters = pd.read_csv(f'CSV/{VERSION} Chapters.csv').set_index('Reference')

reading_times = [f'Time ({word_count} WPM)' for word_count in BibleStats.R_SPEEDS]
sections.drop(reading_times, axis=1, inplace=True)
books.drop(reading_times, axis=1, inplace=True)
chapters.drop(reading_times, axis=1, inplace=True)

sections[f'Time ({WPM_SPEED} WPM)'] = sections['Words'].apply(lambda x: BibleStats.wpm(x, WPM_SPEED))
books[f'Time ({WPM_SPEED} WPM)'] = books['Words'].apply(lambda x: BibleStats.wpm(x, WPM_SPEED))
chapters[f'Time ({WPM_SPEED} WPM)'] = chapters['Words'].apply(lambda x: BibleStats.wpm(x, WPM_SPEED))
sections
```
