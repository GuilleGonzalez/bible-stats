```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

kjv = pd.read_csv('CSV/KJV Books.csv')
esv = pd.read_csv('CSV/ESV Books.csv')
```

## Books

```python
kjv.head()
```

```python
esv.head()
```

```python
df = pd.DataFrame()
df['Book'] = kjv['Book']
df['Chapters'] = abs(kjv['Chapters'] - esv['Chapters'])
df['Verses'] = abs(kjv['Verses'] - esv['Verses'])
df['Words'] = abs(kjv['Words'] - esv['Words'])
df
```

## Sections

```python
kjv = pd.read_csv('CSV/KJV Sections.csv')
esv = pd.read_csv('CSV/ESV Sections.csv')

df = pd.DataFrame()
df['Section'] = kjv['Section']
df['Books'] = abs(kjv['Books'] - esv['Books'])
df['Chapters'] = abs(kjv['Chapters'] - esv['Chapters'])
df['Verses'] = abs(kjv['Verses'] - esv['Verses'])
df['Words'] = abs(kjv['Words'] - esv['Words'])
df
```
