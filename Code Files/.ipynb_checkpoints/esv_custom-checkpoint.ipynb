```python
# DERIVED FROM https://versenotes.org/a-list-of-books-in-the-bible-by-number-of-chapters/ FILES
# RUN 'esv.py' & 'esv_sections.ipynb' FIRST
from bible_stats import BibleStats # My Custom Class for the WPM Function & Constants
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

WPM_SPEED = 300

esv_sections = pd.read_csv('ESV/ESV Sections.csv').set_index('Section')
esv_books = pd.read_csv('ESV/ESV Books.csv').set_index('Book')
esv_chapters = pd.read_csv('ESV/ESV Chapters.csv').set_index('Reference')

def wpm(word_count: int):
    time = word_count / WPM_SPEED
    hrs, mins, secs = int(time // 60), int(time % 60), int(float(time * 60) % 60)
    return f'{hrs}hr {mins}m' if hrs > 0 else f'{mins}m {secs}s'

reading_times = [f'Time ({word_count} WPM)' for word_count in range(BibleStats.MIN_WPM, BibleStats.MAX_WPM, BibleStats.INCREMENT)]
esv_sections.drop(reading_times, axis=1, inplace=True)
esv_books.drop(reading_times, axis=1, inplace=True)
esv_chapters.drop(reading_times, axis=1, inplace=True)

esv_sections[f'Time ({WPM_SPEED} WPM)'] = esv_sections['Words'].apply(lambda x: wpm(x))
esv_books[f'Time ({WPM_SPEED} WPM)'] = esv_books['Words'].apply(lambda x: wpm(x))
esv_chapters[f'Time ({WPM_SPEED} WPM)'] = esv_chapters['Words'].apply(lambda x: wpm(x))
esv_sections
```
