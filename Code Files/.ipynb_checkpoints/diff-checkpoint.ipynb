```python
# ESV DERIVED FROM https://versenotes.org/a-list-of-books-in-the-bible-by-number-of-chapters/ FILES
# RUN 'general.ipynb' & 'esv.py' & 'esv_sections.ipynb' FIRST

import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

esv = pd.read_csv('ESV/ESV Books.csv')
general = pd.read_csv('General/Bible Books (General).csv')
```

```python
esv.head()
```

```python
general.head()
```

```python
df = pd.DataFrame()
df['Book'] = esv['Book']
df['Chapters'] = esv['Chapters'] - general['Chapters']
df['Verses'] = esv['Verses'] - general['Verses']
df['Words'] = esv['Words'] - general['Words']
df
```

```python
esv_sections = pd.read_csv('ESV/ESV Sections.csv')
general_sections = pd.read_csv('General/Bible Sections (General).csv')

df = pd.DataFrame()
df['Section'] = esv_sections['Section']
df['Books'] = esv_sections['Books'] - general_sections['Books']
df['Chapters'] = esv_sections['Chapters'] - general_sections['Chapters']
df['Verses'] = esv_sections['Verses'] - general_sections['Verses']
df['Words'] = esv_sections['Words'] - general_sections['Words']
df
```
